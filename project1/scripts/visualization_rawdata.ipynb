```python
%matplotlib inline
import numpy as np
import pandas as pd 
import os
import random 
import matplotlib.pyplot as plt
from matplotlib.pyplot import ylim, xlim
%matplotlib inline
import seaborn as sns
%load_ext autoreload
%autoreload 2
```

```python
#distribution in train data
plt.style.use('ggplot')

data = pd.read_csv('../data/train.csv')
x_train = data.drop(['Id','Prediction'],axis=1)
y_train = data[['Prediction']]
y_train = y_train.replace('b',-1)
y_train = y_train.replace('s',1)

fig, ax = plt.subplots(6,5, figsize=(16, 18))
ax = ax.flatten()
for i in range(30):
    sns.distplot(x_train.iloc[:,i].values, ax=ax[i])
    ax[i].set_title(x_train.columns[i])
fig.tight_layout(pad=2.0)  
plt.savefig('distribution.jpg')
```

```python
#distribution in test data
data2 = pd.read_csv('../data/test.csv')
x_test = data2.drop(['Id','Prediction'],axis=1)
fig, ax = plt.subplots(6,5, figsize=(16, 18))
ax = ax.flatten()
sns.set_palette('coolwarm')
for i in range(30):
    sns.distplot(x_test.iloc[:,i].values, ax=ax[i])
    ax[i].set_title(x_train.columns[i])
fig.tight_layout(pad=2.0)  
plt.savefig('test_distribution.jpg')
```

```python
#correlation with target
x_train.corrwith(y_train['Prediction']).plot(kind='bar', figsize=(12, 6), title='features correlation with target')
plt.savefig('corr.jpg')
```
